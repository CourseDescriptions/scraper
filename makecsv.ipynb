```python
import json
import os
import pandas as pd
import glob
```

```python
def get_school_csv(schoolname):
    scrapes = glob.glob(f"data/{schoolname}/*.json")
    latest_scrape = max(scrapes, key=os.path.getctime)
    print(latest_scrape)
    with open(latest_scrape) as f:
        data = json.load(f)
    
    df = pd.DataFrame(data["data"])
    df.to_csv(f"{schoolname}_Scrape.csv")
```

```python
get_school_csv("CSU_Monterey_Bay")
```

```python
files = [file for file in os.listdir("data/CSU_Bakersfield/") if file[4] != "_"]

files.sort()

print(files[-1])
```

```python
# Define the directory containing the JSON files
directory = "data/"

# Dictionary to track the most recent file for each school
latest_files = {}
# Process each JSON file
for root, _, files in os.walk(directory):
    for file in files:
        if file.endswith(".json"):
            filepath = os.path.join(root, file)
            school_name = os.path.basename(root)  # Extract school name from folder
            timestamp = file.replace(".json", "")  # Extract timestamp from filename
            if timestamp[4] == "_": continue # avoid old dataa
            
            # Update the latest file for each school
            if school_name not in latest_files or timestamp > latest_files[school_name]["timestamp"]:
                latest_files[school_name] = {"timestamp": timestamp, "filepath": filepath}

allCourses = []
# Process only the most recent files
for entry in latest_files.values():
    filepath = entry["filepath"]
    print(filepath)
    with open(filepath, "r", encoding="utf-8") as file:
        try:
            data = json.load(file)  # Load JSON data
            allCourses.extend(data["data"])
        except json.JSONDecodeError as e:
            print(f"Error reading {filepath}: {e}")

df = pd.DataFrame(allCourses)
df.head(15)
```

```python
df.to_csv("uc_calstate_data.csv")
```
